```python
import control
import matplotlib.pyplot as plt
import numpy as np

# Define Transfer Function: H(s) = (2s^2 + s + 3) / (8s^4 + 12s^3 + 26s^2 + 18s)
num = [2, 1, 3]
den = [8, 12, 26, 18, 0]
sys = control.TransferFunction(num, den)
```

```python
# 1. Step Response Plot
t, y = control.step_response(sys)
plt.figure(figsize=(10, 6))
plt.plot(t, y, linewidth=2)
plt.title('Step Response (Displacement X2)', fontsize=14)
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.grid(True, which='both', linestyle='--')
plt.xlim(0, 20)
plt.savefig('step_response.png')
```

```python
# 2. Pole-Zero Map
plt.figure(figsize=(8, 8))
control.pzmap(sys, plot=True)

# Correct function name: poles
poles = control.poles(sys)

plt.title('Pole-Zero Map', fontsize=14)
plt.grid(True)
plt.xlim(np.real(poles).min() - 1, 1)
plt.ylim(np.imag(poles).min() - 1, np.imag(poles).max() + 1)
plt.savefig('pzmap.png')

print("Analysis complete.")
```
