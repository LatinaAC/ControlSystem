```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the transfer function: G(s) = 1 / (s^4 + 2s^3 + 3s^2 + 2s)
num = [1]
den = [1, 2, 3, 2, 0]
system = ct.tf(num, den)
```

```python
# 2. Calculate the step response for 30 seconds
t = np.linspace(0, 30, 1000)
t, y = ct.step_response(system, T=t)
```

```python
# 3. Calculate step response characteristics
try:
    info = ct.step_info(system, T=t)
except Exception:
    info = "Characteristics not applicable for non-converging system."
```

```python

# 4. Calculate poles and zeros
poles = ct.poles(system)
zeros = ct.zeros(system)
```

```python
# 5. Create visualizations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Step Response Plot
ax1.plot(t, y, color='blue', linewidth=2)
ax1.set_title('System Step Response', fontsize=14)
ax1.set_xlabel('Time (s)', fontsize=12)
ax1.set_ylabel('Angular Displacement θ2', fontsize=12)
ax1.grid(True, which='both', linestyle='--', alpha=0.5)

# Pole-Zero Map
ct.pole_zero_plot(system, ax=ax2)
ax2.set_title('Pole-Zero Map', fontsize=14)
ax2.grid(True, which='both', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
