```python
import numpy as np
import matplotlib.pyplot as plt
import control

# Define the transfer function H(s) = (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
num = [1, 1]
den = [1, 4, 7, 6, 3]
sys = control.TransferFunction(num, den)
```

```python

# Calculate poles and zeros
poles = control.poles(sys)
zeros = control.zeros(sys)
```

```python

# Calculate step response metrics
t = np.linspace(0, 20, 1000)
T, yout = control.step_response(sys, T=t)
info = control.step_info(sys, T=T)
steady_state_value = control.dcgain(sys)
```

```python

# Plotting Results
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))
ax1.plot(T, yout, linewidth=2)
ax1.axhline(steady_state_value, color='r', linestyle='--')
ax1.set_title('Step Response')
ax1.grid(True)

plt.sca(ax2)
control.pzmap(sys, plot=True)
ax2.set_title('Pole-Zero Map')
plt.tight_layout()
plt.show()
```
